```python
# --- config ---------------------------------------------------------------
# Define each analysis run here. Each dictionary groups all the information
# for a single bar group in the final plot.



# csa moment 1, order history NOT included in order generation prompt
RUN_CONFIGS = [
    {
        "dir": "/mnt/i/diplo/sam-exp087-csa",
        "label": "sam-exp087-csa baseline",
        "phase": "F1907M",
        "power": "GERMANY",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp087-csa-r",
        "label": "sam-exp087-csa-r baseline (repeat)",
        "phase": "F1907M",
        "power": "GERMANY",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp087-csa2",
        "label": "sam-exp087-csa2 supports explained v2 nshot3",
        "phase": "F1907M",
        "power": "GERMANY",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp087-csa3",
        "label": "sam-exp087-csa3 supports explained v2 nshot3 + condensed possible moves",
        "phase": "F1907M",
        "power": "GERMANY",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp087-csa4",
        "label": "sam-exp087-csa4 condensed possible moves",
        "phase": "F1907M",
        "power": "GERMANY",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp087-csa5",
        "label": "sam-exp087-csa5 supports explained v2 nshot6",
        "phase": "F1907M",
        "power": "GERMANY",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp087-csa6",
        "label": "sam-exp087-csa6 supports explained v2 nshot12",
        "phase": "F1907M",
        "power": "GERMANY",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp087-csa7",
        "label": "sam-exp087-csa7 supports explained v2 nshot6 + condensed possible moves",
        "phase": "F1907M",
        "power": "GERMANY",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp087-csa8",
        "label": "sam-exp087-csa8 supports explained v2 nshot12 + condensed possible moves",
        "phase": "F1907M",
        "power": "GERMANY",
    },
]

# csa moment 1, order history IS included in order generation prompt
RUN_CONFIGS = [
    {
        "dir": "/mnt/i/diplo/sam-exp094-csa",
        "label": "sam-exp094-csa baseline",
        "phase": "F1907M",
        "power": "GERMANY",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp094-csa-r",
        "label": "sam-exp094-csa-r baseline (repeat)",
        "phase": "F1907M",
        "power": "GERMANY",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp094-csa2",
        "label": "sam-exp094-csa2 supports explained v2 nshot3",
        "phase": "F1907M",
        "power": "GERMANY",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp094-csa3",
        "label": "sam-exp094-csa3 supports explained v2 nshot3 + condensed possible moves",
        "phase": "F1907M",
        "power": "GERMANY",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp094-csa4",
        "label": "sam-exp094-csa4 condensed possible moves",
        "phase": "F1907M",
        "power": "GERMANY",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp094-csa5",
        "label": "sam-exp094-csa5 supports explained v2 nshot6",
        "phase": "F1907M",
        "power": "GERMANY",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp094-csa6",
        "label": "sam-exp094-csa6 supports explained v2 nshot12",
        "phase": "F1907M",
        "power": "GERMANY",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp094-csa7",
        "label": "sam-exp094-csa7 supports explained v2 nshot6 + condensed possible moves",
        "phase": "F1907M",
        "power": "GERMANY",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp094-csa8",
        "label": "sam-exp094-csa8 supports explained v2 nshot12 + condensed possible moves",
        "phase": "F1907M",
        "power": "GERMANY",
    },
]

# csa moment 2, order history IS included in order generation prompt
_RUN_CONFIGS = [
    {
        "dir": "/mnt/i/diplo/sam-exp093-csa",
        "label": "sam-exp093-csa baseline",
        "phase": "F1917M",
        "power": "AUSTRIA",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp093-csa-r",
        "label": "sam-exp093-csa-r baseline (repeat)",
        "phase": "F1917M",
        "power": "AUSTRIA",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp093-csa2",
        "label": "sam-exp093-csa2 supports explained v2 nshot3",
        "phase": "F1917M",
        "power": "AUSTRIA",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp093-csa3",
        "label": "sam-exp093-csa3 supports explained v2 nshot3 + condensed possible moves",
        "phase": "F1917M",
        "power": "AUSTRIA",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp093-csa4",
        "label": "sam-exp093-csa4 condensed possible moves",
        "phase": "F1917M",
        "power": "AUSTRIA",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp093-csa5",
        "label": "sam-exp093-csa5 supports explained v2 nshot6",
        "phase": "F1917M",
        "power": "AUSTRIA",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp093-csa6",
        "label": "sam-exp093-csa6 supports explained v2 nshot12",
        "phase": "F1917M",
        "power": "AUSTRIA",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp093-csa7",
        "label": "sam-exp093-csa7 supports explained v2 nshot6 + condensed possible moves",
        "phase": "F1917M",
        "power": "AUSTRIA",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp093-csa8",
        "label": "sam-exp093-csa8 supports explained v2 nshot12 + condensed possible moves",
        "phase": "F1917M",
        "power": "AUSTRIA",
    },
]

# csa moment 2, order history NOT included in order generation prompt
_RUN_CONFIGS = [
    {
        "dir": "/mnt/i/diplo/sam-exp095-csa",
        "label": "sam-exp095-csa baseline",
        "phase": "F1917M",
        "power": "AUSTRIA",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp095-csa-r",
        "label": "sam-exp095-csa-r baseline (repeat)",
        "phase": "F1917M",
        "power": "AUSTRIA",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp095-csa2",
        "label": "sam-exp095-csa2 supports explained v2 nshot3",
        "phase": "F1917M",
        "power": "AUSTRIA",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp095-csa3",
        "label": "sam-exp095-csa3 supports explained v2 nshot3 + condensed possible moves",
        "phase": "F1917M",
        "power": "AUSTRIA",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp095-csa4",
        "label": "sam-exp095-csa4 condensed possible moves",
        "phase": "F1917M",
        "power": "AUSTRIA",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp095-csa5",
        "label": "sam-exp095-csa5 supports explained v2 nshot6",
        "phase": "F1917M",
        "power": "AUSTRIA",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp095-csa6",
        "label": "sam-exp095-csa6 supports explained v2 nshot12",
        "phase": "F1917M",
        "power": "AUSTRIA",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp095-csa7",
        "label": "sam-exp095-csa7 supports explained v2 nshot6 + condensed possible moves",
        "phase": "F1917M",
        "power": "AUSTRIA",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp095-csa8",
        "label": "sam-exp095-csa8 supports explained v2 nshot12 + condensed possible moves",
        "phase": "F1917M",
        "power": "AUSTRIA",
    },
]


# csa moment 3, order history NOT included in order generation prompt
RUN_CONFIGS = [
    {
        "dir": "/mnt/i/diplo/sam-exp096-csa",
        "label": "sam-exp096-csa baseline",
        "phase": "S1912M",
        "power": "FRANCE",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp096-csa-r",
        "label": "sam-exp096-csa-r baseline (repeat)",
        "phase": "S1912M",
        "power": "FRANCE",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp096-csa2",
        "label": "sam-exp096-csa2 supports explained v2 nshot3",
        "phase": "S1912M",
        "power": "FRANCE",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp096-csa3",
        "label": "sam-exp096-csa3 supports explained v2 nshot3 + condensed possible moves",
        "phase": "S1912M",
        "power": "FRANCE",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp096-csa4",
        "label": "sam-exp096-csa4 condensed possible moves",
        "phase": "S1912M",
        "power": "FRANCE",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp096-csa5",
        "label": "sam-exp096-csa5 supports explained v2 nshot6",
        "phase": "S1912M",
        "power": "FRANCE",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp096-csa6",
        "label": "sam-exp096-csa6 supports explained v2 nshot12",
        "phase": "S1912M",
        "power": "FRANCE",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp096-csa7",
        "label": "sam-exp096-csa7 supports explained v2 nshot6 + condensed possible moves",
        "phase": "S1912M",
        "power": "FRANCE",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp096-csa8",
        "label": "sam-exp096-csa8 supports explained v2 nshot12 + condensed possible moves",
        "phase": "S1912M",
        "power": "FRANCE",
    },
]

# csa moment 3, order history IS included in order generation prompt
_RUN_CONFIGS = [
    {
        "dir": "/mnt/i/diplo/sam-exp097-csa",
        "label": "sam-exp097-csa baseline",
        "phase": "S1912M",
        "power": "FRANCE",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp097-csa-r",
        "label": "sam-exp097-csa-r baseline (repeat)",
        "phase": "S1912M",
        "power": "FRANCE",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp097-csa2",
        "label": "sam-exp097-csa2 supports explained v2 nshot3",
        "phase": "S1912M",
        "power": "FRANCE",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp097-csa3",
        "label": "sam-exp097-csa3 supports explained v2 nshot3 + condensed possible moves",
        "phase": "S1912M",
        "power": "FRANCE",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp097-csa4",
        "label": "sam-exp097-csa4 condensed possible moves",
        "phase": "S1912M",
        "power": "FRANCE",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp097-csa5",
        "label": "sam-exp097-csa5 supports explained v2 nshot6",
        "phase": "S1912M",
        "power": "FRANCE",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp097-csa6",
        "label": "sam-exp097-csa6 supports explained v2 nshot12",
        "phase": "S1912M",
        "power": "FRANCE",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp097-csa7",
        "label": "sam-exp097-csa7 supports explained v2 nshot6 + condensed possible moves",
        "phase": "S1912M",
        "power": "FRANCE",
    },
    {
        "dir": "/mnt/i/diplo/sam-exp097-csa8",
        "label": "sam-exp097-csa8 supports explained v2 nshot12 + condensed possible moves",
        "phase": "S1912M",
        "power": "FRANCE",
    },
]



# --- imports --------------------------------------------------------------
from pathlib import Path
import json, numpy as np, matplotlib.pyplot as plt, textwrap as tw

# --- helpers --------------------------------------------------------------
def get_stats_for_phase(exp_dir: Path, target_phase: str, target_power: str):
    """
    Return arrays of move-success/unit and support-success/unit for a
    specific power in a specific phase, with one data point per game file.
    """
    move_ratios, support_ratios = [], []
    for jpath in exp_dir.glob("runs/run_*/lmvsgame.json"):
        with jpath.open(encoding="utf-8") as fh:
            data = json.load(fh)

        # Find the specific phase in the game log
        phase_data = next(
            (p for p in data.get("phases", [])
             if (p.get("name") or p.get("state", {}).get("name", "")) == target_phase),
            None
        )

        if not phase_data:
            continue  # Target phase not found in this game file

        units = phase_data["state"]["units"].get(target_power, [])
        n_units = len(units)
        if n_units == 0:
            continue  # Target power eliminated or has no units in this phase

        orders = phase_data.get("order_results", {}).get(target_power, {})
        succ_moves    = sum(1 for o in orders.get("move",    []) if o["result"] == "success")
        succ_supports = sum(1 for o in orders.get("support", []) if o["result"] == "success")

        move_ratios.append(succ_moves / n_units)
        support_ratios.append(succ_supports / n_units)

        del data

    return np.array(move_ratios), np.array(support_ratios)

def bootstrap_ci(samples, iters=10_000, alpha=0.05):
    """Return mean, lower-CI, upper-CI for the bootstrap of the mean."""
    if samples.size == 0:
        return np.nan, np.nan, np.nan
    rng = np.random.default_rng()
    means = rng.choice(samples, size=(iters, samples.size), replace=True).mean(axis=1)
    lower, upper = np.percentile(means, [100*alpha/2, 100*(1-alpha/2)])
    return samples.mean(), lower, upper

# --- aggregate ------------------------------------------------------------
summary = {}
for config in RUN_CONFIGS:
    exp_dir = config["dir"]
    move, sup = get_stats_for_phase(Path(exp_dir), config["phase"], config["power"])
    combined = move + sup
    m_mean, m_low, m_up = bootstrap_ci(move)
    s_mean, s_low, s_up = bootstrap_ci(sup)
    c_mean, c_low, c_up = bootstrap_ci(combined)
    # Use the directory as a unique key for the summary stats
    summary[exp_dir] = dict(
        move_mean=m_mean, move_low=m_low, move_up=m_up,
        support_mean=s_mean, support_low=s_low, support_up=s_up,
        combined_mean=c_mean, combined_low=c_low, combined_up=c_up,
        n_samples=move.size,
    )

# --- textual report -------------------------------------------------------
print("Analysis Report\n" + "="*50)
for config in RUN_CONFIGS:
    stats = summary[config["dir"]]
    print(
        f"{config['label']} (n={stats['n_samples']} games)\n"
        f"  Analysis Target: Power '{config['power']}' in Phase '{config['phase']}'\n"
        f"  move_success/unit   = {stats['move_mean']:.3f}  "
        f"[{stats['move_low']:.3f}, {stats['move_up']:.3f}]\n"
        f"  support_success/unit= {stats['support_mean']:.3f}  "
        f"[{stats['support_low']:.3f}, {stats['support_up']:.3f}]\n"
        f"  move+support/unit   = {stats['combined_mean']:.3f}  "
        f"[{stats['combined_low']:.3f}, {stats['combined_up']:.3f}]\n"
    )

# --- plot -----------------------------------------------------------------
labels         = [tw.fill(c['label'], width=38) for c in RUN_CONFIGS]
move_means     = np.array([summary[c["dir"]]["move_mean"]     for c in RUN_CONFIGS])
support_means  = np.array([summary[c["dir"]]["support_mean"]  for c in RUN_CONFIGS])
combined_means = np.array([summary[c["dir"]]["combined_mean"] for c in RUN_CONFIGS])

move_err = np.vstack([
    move_means - np.array([summary[c["dir"]]["move_low"] for c in RUN_CONFIGS]),
    np.array([summary[c["dir"]]["move_up"] for c in RUN_CONFIGS]) - move_means,
])
sup_err = np.vstack([
    support_means - np.array([summary[c["dir"]]["support_low"] for c in RUN_CONFIGS]),
    np.array([summary[c["dir"]]["support_up"] for c in RUN_CONFIGS]) - support_means,
])
comb_err = np.vstack([
    combined_means - np.array([summary[c["dir"]]["combined_low"] for c in RUN_CONFIGS]),
    np.array([summary[c["dir"]]["combined_up"] for c in RUN_CONFIGS]) - combined_means,
])

y = np.arange(len(RUN_CONFIGS))
h = 0.25

fig, ax = plt.subplots(figsize=(10, max(4, len(labels)*0.8)))
ax.barh(y - h,   move_means,     h, xerr=move_err, label="Move success / unit")
ax.barh(y,       support_means,  h, xerr=sup_err,  label="Support success / unit")
ax.barh(y + h,   combined_means, h, xerr=comb_err, label="Move+Support success / unit")

ax.set_xlabel("Successes per Unit (for specified power/phase)")
ax.set_yticks(y)
ax.set_yticklabels(labels)
ax.invert_yaxis()
ax.set_title("Per-phase Success Rates by Experiment (95% CIs)")

# ── legend outside the axes ───────────────────────────────────────────────
fig.legend(
    loc="upper center",
    bbox_to_anchor=(0.5, -0.04),   # centred, just below the axes
    ncol=3,                        # three labels side-by-side
    frameon=False,
)

# reserve a little bottom margin for the legend
plt.tight_layout(rect=[0, 0.05, 1, 1])   # left, bottom, right, top (0-1)

plt.show()
```
